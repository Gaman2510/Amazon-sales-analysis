<p style="text-align:center;color:black;font-size:35px;"><b>Loading Amazon sales dataset</b></p>

```python
import numpy as np
import pandas as pd
data=pd.read_csv("C:/Users/Admin/Desktop/Gaman/amazon.csv")
df=pd.DataFrame(data)
df.head()
```

<p style="text-align:center;color:black;font-size:35px;"><b>Cleaning the dataset</b></p>

<p style="text-align:left;color:black;font-size:25px;"><b>Finding null values,duplicates and the datatypes </b></p>

```python
print(df.isnull().sum())
print(df.duplicated().sum())
print(df.dtypes)
```

<p style="text-align:left;color:black;font-size:25px;"><b>Treating the null values</b></p>

```python
import warnings
warnings.filterwarnings('ignore')
df['rating_count']=pd.to_numeric(df['rating_count'],errors='coerce')
df['rating_count'].fillna(df['rating_count'].median(),inplace=True)
```

<p style="text-align:left;color:black;font-size:25px;"><b>Datatype convertion,removing unwanted symbols and extra spaces</b></p>

```python
for i in ['product_name','category','about_product','user_name','review_title','review_content']:
    df[i]=df[i].str.strip()
```

```python
df['discounted_price']=df['discounted_price'].astype(str).str.replace('₹','',regex=False).str.replace(",","")
```

```python
df['actual_price']=df['actual_price'].astype(str).str.replace('₹','',regex=False).str.replace(",","")
df['discount_percentage']=df['discount_percentage'].astype(str).str.replace('%','',regex=False).str.replace(",","")
```

```python
for i in  ['discount_percentage','rating_count']:
    df[i]=df[i].astype(int)
```

```python
df['rating']=df['rating'].astype(str).str.replace('|','0',regex=False)
```

```python
for i in  ['actual_price','discounted_price','rating']:
    df[i]=df[i].astype(float)
```

```python
df['rating']=df['rating'].replace(0.0,np.nan)
```

```python
df['rating']=df['rating'].fillna(df['rating'].median())
```

```python
df['category']=df['category'].str.lower()
```

```python
df['category']=df['category'].str.replace('|',',')
```

```python
for i in ['product_name','about_product','user_name','review_title','review_content']:
    df[i]=df[i].str.lower()
```

<p style="text-align:left;color:black;font-size:25px;"><b>Removing unwanted columns</b></p>

```python
df=df.drop(columns=['about_product','user_name','review_title','review_content','img_link'])
```

<p style="text-align:center;color:black;font-size:35px;"><b>Performing SQL operations on the cleaned Amazon dataset</b></p>

```python
import sqlalchemy as db
from sqlalchemy import create_engine
import pymysql
```

```python
cl=pd.read_csv("amazon_cleaned.csv")
```

```python
engine=db.create_engine("sqlite:///mydb.db")
engine
```

```python
cl.to_sql("amazon_cleaned",con=engine,if_exists="replace",index=False)
```

```python
rows=pd.read_sql("select*from amazon_cleaned",engine).head()
rows
```

<p style="text-align:left;color:black;font-size:25px;"><b>List all the products ratings above 4.5</b></p>

```python
ratings=pd.read_sql("select*from amazon_cleaned where(rating>4.5)",engine).head()
ratings
```

<p style="text-align:left;color:black;font-size:25px;"><b>Count the total number of products availabel in the dataset</b></p>

```python
count=pd.read_sql("select count() as Total_Products from amazon_cleaned",engine)
count
```

<p style="text-align:left;color:black;font-size:25px;"><b>Show the names and discounted prices of products that offer more than a 50% discount</b></p>

```python
discount=pd.read_sql("select product_name,discounted_price from amazon_cleaned where(discount_percentage>50)",engine).head()
discount
```

<p style="text-align:left;color:black;font-size:25px;"><b>Display the product IDs and categories for products priced below 200</b></p>

```python
Id=pd.read_sql("select product_id,category from amazon_cleaned where(actual_price<200)",engine).head()
Id
```

<p style="text-align:left;color:black;font-size:25px;"><b>Find the average rating for each unique discount percentage</b></p>

```python
avg_rating=pd.read_sql("select avg(rating)as Average_ratings,discount_percentage from amazon_cleaned group by discount_percentage",engine).head()
avg_rating
```

<p style="text-align:left;color:black;font-size:25px;"><b>List the top 5 products with the highest number of ratings</b></p>

```python
high_rating=pd.read_sql("select * from amazon_cleaned group by rating_count order by rating_count desc",engine).head()
high_rating
```

<p style="text-align:left;color:black;font-size:25px;"><b>Retrieve all the products that belong to the 'USB' category</b></p>

```python
usb=pd.read_sql("select*from amazon_cleaned where category like '%usbcables%'",engine).head()
usb
```

<p style="text-align:left;color:black;font-size:25px;"><b>Find the products that have highest discount percentage</b></p>

```python
print("1.Find the products that have highest discount percentage")
high_disc=pd.read_sql("select*from amazon_cleaned where discount_percentage=(select max(discount_percentage) from amazon_cleaned)",engine)
high_disc
```

<p style="text-align:left;color:black;font-size:25px;"><b>Calculate the average discounted price of products rated above 4.0, grouped by category</b></p>

```python
avg_discp=pd.read_sql("select avg(discounted_price),category,rating from amazon_cleaned where(rating>4) group by category",engine).head()
avg_discp
```

<p style="text-align:left;color:black;font-size:25px;"><b>Find products where the actual price is more than three times the discounted price</b></p>

```python
disc=pd.read_sql("select*from amazon_cleaned where(actual_price>3*discounted_price)",engine).head()
disc
```
